```python
import pandas as pd
import matplotlib.pyplot as plt

stars_w_gravity_df = pd.read_csv('https://drive.google.com/uc?id=1hf_L-8BAQTp4bSZdlMs8jQPnGpmN9Siw') #https://drive.google.com/file/d/1hf_L-8BAQTp4bSZdlMs8jQPnGpmN9Siw/view?usp=drive_link

stars_w_gravity_df.head(10)
```

```python
distance_filtered_stars = stars_w_gravity_df.drop(stars_w_gravity_df[stars_w_gravity_df['Distance'] > 100].index)
len(distance_filtered_stars)
```

```python
distance_filtered_stars.head(10)
```

```python
gravity_filtered_stars = distance_filtered_stars.drop(distance_filtered_stars[distance_filtered_stars['Gravity'] < 150].index)
gravity_filtered_stars = gravity_filtered_stars.drop(gravity_filtered_stars[gravity_filtered_stars['Gravity'] > 350].index)

len(gravity_filtered_stars)
```
